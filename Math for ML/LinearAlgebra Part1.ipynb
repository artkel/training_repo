### Matrix Tensors

```python
X.size
```

```python
X.shape
```

```python
X.T
```

```python
X_gram = np.dot(X, X.T)
```

```python
X_gram
```

```python
np.linalg.matrix_rank(X)
```

```python
np.linalg.matrix_rank(X_gram)
```

```python
np.linalg.inv(X_gram)
```

```python
np.dot(X_gram, np.linalg.inv(X_gram))
```
