```python
from scipy.integrate import quad
```

From the slides: $$ \int_1^2 \frac{x}{2} dx = \frac{3}{4} $$

```python
def g(x):
    return x/2
```

```python
quad(g,1,2)
```

The second output of `quad` is an estimate of the absolute error of the integral, which in this case is essentially zero.

```python
def f(x):
    return x*2
```

```python
quad(f,3,4)
```
