```python


classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

```python
classifier(
    "Lionel Messi is the greatest footbal player that has ever kicked the ball",
    candidate_labels=["soccer", "politics", "happiness"],
)
```

```python
generator = pipeline("text-generation", model="inception-mbzuai/jais-13b")
generator("In this course, we will teach you how to", num_return_sequences=3, max_length=120)
```

```python
from transformers import pipeline

unmasker = pipeline("fill-mask", model="bert-base-cased")
unmasker("This course will teach you all about [MASK] models.", top_k=2)
```

```python
ner = pipeline("ner", grouped_entities=True)
ner("Lionel Messi is a professional soccer player, currently playing for Inter Miami in the USA")
```

```python
translator = pipeline("translation", model="Helsinki-NLP/opus-mt-en-de")
translator("I würde gerne mehr über die Implementierung von Machine Learning Modellen erfahren!")
```
